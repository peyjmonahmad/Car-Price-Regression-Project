```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import random
import os
import requests
from bs4 import BeautifulSoup

chromedriver = "/Applications/chromedriver" # path to the chromedriver executable
os.environ["webdriver.chrome.driver"] = chromedriver

driver = webdriver.Chrome(chromedriver)
```

```python
def looping(url):
    
    price= 'div.text-gray-base.text-size-30.margin-right-auto'
    general_tag = 'div.text-sm.text-gray.text-uppercase'
    general_val = 'div.text-base.text-bold'
    specs_button = '//button[text()="View Features and Specifications"]'
    list_of_specs = 'li.list-bordered.list-condensed'
    car_details = '//a[contains(@href, "/cars-for-sale/vehicledetails.xhtml?listingId")]'
        
    os.environ["webdriver.chrome.driver"] = chromedriver
    driver = webdriver.Chrome(chromedriver)
        
    driver.get(url)
      
    all_cars = []
    
    #Web element for each Car
    xpaths = driver.find_elements_by_xpath(car_details) 
    xpaths = xpaths[0:100]
  
 
  
    ##Close Ad survey's if it pops up.
    try:
        #Random sleep times distributed to difuse web blocking
        time.sleep(random.randint(1,5))
        driver.find_element_by_css_selector('a.acsCloseButton.acsAbandonButton').click()
            
    except:
        pass
        
    try:
        time.sleep(random.randint(1,3))
        driver.find_element_by_css_selector('a.acsInviteButton.acsDeclineButton').click()
        
    except:
        pass
    
    try:
        
        driver.find_element_by_css_selector('div.caoEngagementClosure').click()
    
    except:
        pass
    

    #Loops through each unique link
    for element in range(len(xpaths)):
        
        #Initialize dict for each individual car
        car_dict = {}
        
        try:
            time.sleep(random.randint(1,3))
            driver.find_element_by_css_selector('a.acsCloseButton.acsAbandonButton').click()
        except:
            pass
            
        try:
            time.sleep(random.randint(1,3))
            driver.find_element_by_css_selector('a.acsInviteButton.acsDeclineButton').click()
            
        except:
            pass
        
        try:
            driver.find_element_by_css_selector('div.caoEngagementClosure').click()
            
        except:
            pass
            
            
        try:
            xpaths[element].click()
                
        except:
            continue
            
        # general feature names : Mileage, Drive Type, Engine, MPG, Transmission, Fuel Type, Interior, Exterior  
        name = driver.find_elements_by_css_selector('h1.text-size-sm-40.text-size-30.text-light.display-inline.margin-right-2')
        
        
        name_dict = {}
        
        for i in name:
            name_dict['Title'] = (i.text)
        
        car_dict.update(name_dict)
            
        
        #Price for each car
        prices = driver.find_elements_by_css_selector('div.text-gray-base.text-size-30.margin-right-auto')
            
        price_dict = {}
            
        for i in prices:
            price_dict['Price'] = (i.text)
        car_dict.update(price_dict)
            
        gen_names = []
            
        general = driver.find_elements_by_css_selector(general_tag)
            
        for i in general:
            gen_names.append(i.text)
            
            
        # General feature values   
        gen_nums = []
        gen1 =driver.find_elements_by_css_selector(general_val)
                
        for i in gen1:
            gen_nums.append(i.text)
                    
        # Make zipped Dict of both combined
        general_dict = dict(zip(gen_names,gen_nums))
                
        car_dict.update(general_dict)
                
        time.sleep(random.randint(1,5))
                
        try: 
            # click on View Features and Specs button
            driver.find_element_by_xpath(specs_button).click()
            
            time.sleep(random.randint(1,5))
                
        except:
            continue
            
        ##Takes relevant info from Features and Specs button
        specs_list = driver.find_elements_by_css_selector(list_of_specs)
                
        ## Makes a dict of Spec items as Key, Spec value as Value
        specs_items= {}
        
        # Take first 20 rows of Specs & Features Table
        for i in (specs_list[:20]):
            i = i.text
            i = i.split('\n')
            
            try:
                specs_items[i[0]] = i[1]
            except:
                pass
            
        #Updates dict by adding new dict    
        car_dict.update(specs_items)
        
        #Appends total dict to all_Cars list
        new_cars.append(car_dict)
                
        time.sleep(random.randint(1,5))
                
        #Go back to home page
        try:
            driver.get(url)
        except:
            continue
        
        xpaths = driver.find_elements_by_xpath(car_details)
                
    return new_cars
```
